## Задание (вариант2)
Дан текст, разделенный на абзацы. Необходимо построить обратный индекс этого текста, показывающий в каком абзаце и на какой позиции хранится слово. Пользователь вводит два слова. Необходимо определить, находятся ли эти два слова в соседних абзацах. Если слова находятся в одном абзаце, необходимо найти расстояние между ними.

Следующая функция отдаст список абзацев, в котором каждый абзац представляет собой список слов, и список индексов (внутри этого списка для каждого абзаца заводится список, в котором для каждого слова также заводится список из 2 элементов - номера абзаца и индекса слова в этом абзаце):

```python
def abz_splitter (filename):
    index_list = []
    with open(filename, encoding = 'utf-8') as ff:
        text = ff.read()
        abz_splitted = text.split('\n') #список абзацев текста
        prep = ['.', ',', ':', ';', '?', '!', '(', ')']
        all_abz_without_prep = []
        for abz in abz_splitted:
            abz_text = ''.join(j for j in abz if not j in prep)  #удаление знаков препинания, приклеившихся к словам
            abz_l = abz_text.split()  #список слов данного абзаца
            all_abz_without_prep.append(abz_l)
            index_l=[] 
            for w in abz_l:  #для каждого слова заводим пустой список, в который добавим номер абзаца и индекс слова в этом абзаце
                index_list1 = []
                index_list1.append(abz_splitted.index(abz))
                index_list1.append(abz_l.index(w))
                index_l.append(index_list1)  #добавляем списки для отдельных слов в  список, хранящий все индексы абзаца
            index_list.append(index_l) #большой список, который состоит из подсписков, индекс каждого из которых - номер абзаца, те в свою очередь состоят из двухэлементных подсписков - номер абзаца, порядок слова
                
    return all_abz_without_prep, index_list

abz_splitter('first_text.txt')
    
```

Функция words_finder() работает с 2 введенными словами: если оба слова есть в исходно тексте, в список checker записываются два значения True. Если в этом списке в итоге два значения и оба True, возвращаются индексы каждого слова: номер абзаца и номер слова в абзаце

```python
def words_finder(filename, word1, word2):
    all_abz_without_prep, index_list = abz_splitter (filename) #в файле находится исходный текст
    checker=[]
    counter=[]
    for elem in all_abz_without_prep: #перебираем каждый список слов каждого абзаца
        if word1 in elem:
            checker.append(True)
            counter.append(index_list[all_abz_without_prep.index(elem)][elem.index(word1)])
        if word2 in elem:
            checker.append(True)
            counter.append(index_list[all_abz_without_prep.index(elem)][elem.index(word2)])
        if checker[0] == True and checker[1] == True:
            return counter
        else:
            answer = 'Какого-то слова нет в тексте'
            return answer
    
            
```

Пример работы кода на двух введенных пользователем словах:

```python
a=input(str('Первое слово '))
b=input(str('Второе слово '))
words_finder('first_text.txt', a, b)
```

Дополненная предыдущая функция, возвращает индексы каждого из слов, если они встретились в тексте, а также информацию об их расположении (если встретились в одном абзаце - расстояние между словами, если в соседних - сообщает об этом, если не в соседних - также сообщает об этом)

```python
def words_finder(filename, word1, word2):
    all_abz_without_prep, index_list = abz_splitter (filename) #в файле находится исходный текст
    checker=[]
    counter=[]
    for elem in all_abz_without_prep: #перебираем каждый список слов каждого абзаца
        if word1 in elem:   #если слово1 и слово2 есть в каком-то из абзацев, в список checker записывается значение True, в список counter добавляются индексы этих слов (в виде словаря [номер абзаца, номер слова])
            checker.append(True)
            counter.append(index_list[all_abz_without_prep.index(elem)][elem.index(word1)])
        if word2 in elem:
            checker.append(True)
            counter.append(index_list[all_abz_without_prep.index(elem)][elem.index(word2)])
        if checker[0] == True and checker[1] == True:
            if counter[0][0] == counter [1][0]:  #если слова находятся в одном абзаце
                diff = abs(counter[1][1] - counter[0][1])
                return counter, diff  #вернутся индексы слов в тексте и расстояние между словами
            else:
                some = counter[0][0] - counter [1][0]  #расстояние между абзацами 1 (слова в соседних абзацах)
                if abs(some) == 1:
                    r='Слова находятся в соседних абзацах'
                    return r
                else:  #расстояние между абзацами больше одного
                    r='Слова не находятся ни в одном абзаце, ни в соседних'
                    return r
            return counter
        else:  
            answer = 'Какого-то слова нет в тексте'
            return answer
    
```

Пример работы кода:

```python
a=input(str('Первое слово '))
b=input(str('Второе слово '))
words_finder('first_text.txt', a, b)
```
